# Sinusoidal Signal

```python
import numpy as np
import matplotlib.pyplot as plt

frequency = 5
sampling_rate = 1000
duration = 1

# Time vector
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)
# Sine wave signal
sine_wave = np.sin(2 * np.pi * frequency * t)
```

```python
# Plotting the sine wave
plt.figure(figsize=(8, 2))
plt.plot(t, sine_wave)
plt.title("5Hz Sine Wave")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
n = len(sine_wave)
fft_signal = np.fft.fft(sine_wave)
fft_freq = np.fft.fftfreq(n, d=1/sampling_rate)

positive_freqs = fft_freq[:n//2]
fft_magnitude = np.abs(fft_signal[:n//2])

plt.figure(figsize=(8, 2))
plt.plot(positive_freqs, fft_magnitude)
plt.title('Frequency Spectrum (DTFT approximation using FFT)')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 2))
plt.stem(positive_freqs, fft_magnitude, basefmt=" ", use_line_collection=True)
plt.title('Discrete Frequency Spectrum (DFT)')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

# Composite Signal

```python
import numpy as np
import matplotlib.pyplot as plt

fs = 1000
t = np.arange(0, 1, 1/fs)
f1 = 5
f2 = 50
A1 = 1
A2 = 0.5

signal1 = A1 * np.sin(2 * np.pi * f1 * t)
signal2 = A2 * np.sin(2 * np.pi * f2 * t)
composite_signal = signal1 + signal2


plt.figure(figsize=(8, 2))
plt.plot(t, composite_signal, label='Composite Signal', color='blue')
plt.title('Time-Domain Waveform of Composite Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()
plt.xlim(0, 1)
plt.show()





```

```python
N = 1024
frequencies = np.linspace(-fs/2, fs/2, N)
DTFT = np.fft.fftshift(np.fft.fft(composite_signal, N))

plt.figure(figsize=(8, 2))
plt.plot(frequencies, np.abs(DTFT), color='red')
plt.title('Continuous Frequency Spectrum (DTFT)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid()
plt.xlim(-fs/2, fs/2)
plt.show()
```

```python
DFT = np.fft.fft(composite_signal)

frequencies_dft = np.fft.fftfreq(len(DFT), d=1/fs)

plt.figure(figsize=(8, 2))
plt.stem(frequencies_dft[:len(frequencies_dft)//2], np.abs(DFT)[:len(DFT)//2], use_line_collection=True)
plt.title('Discrete Frequency Spectrum (DFT)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid()
plt.xlim(0, fs/2)
plt.show()
```

# Exponentially Decaying Signal


```python
import numpy as np
import matplotlib.pyplot as plt

fs = 1000
t = np.arange(0, 2, 1/fs)
tau = 0.5

exponential_signal = np.exp(-t / tau)

plt.figure(figsize=(8, 2))
plt.plot(t, exponential_signal, label='Exponential Decay', color='blue')
plt.title('Time-Domain Waveform of Exponentially Decaying Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()
plt.xlim(0, 2)
plt.show()
```

```python
N = 1024
frequencies = np.linspace(-fs/2, fs/2, N)
DTFT = np.fft.fftshift(np.fft.fft(exponential_signal, N))

plt.figure(figsize=(8, 2))
plt.plot(frequencies, np.abs(DTFT), color='red')
plt.title('Continuous Frequency Spectrum (DTFT)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid()
plt.xlim(-fs/2, fs/2)
plt.show()
```

```python
DFT = np.fft.fft(exponential_signal)

frequencies_dft = np.fft.fftfreq(len(DFT), d=1/fs)

plt.figure(figsize=(8, 2))
plt.stem(frequencies_dft[:len(frequencies_dft)//2], np.abs(DFT)[:len(DFT)//2], use_line_collection=True)
plt.title('Discrete Frequency Spectrum (DFT)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid()
plt.xlim(0, fs/2)
plt.show()
```

# Rectangular Function

```python

fs = 1000
T = 1
t = np.linspace(-T/2, T/2, int(fs*T), endpoint=False)


rect_width = 0.2
x = np.where(np.abs(t) <= rect_width/2, 1, 0) # Rectangular function

plt.figure(figsize=(8, 2))
plt.plot(t, x)
plt.title('Time-Domain Waveform of Rectangular Function')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python

N=2048
frequencies = np.fft.fftfreq(N, 1/fs)
x_dtft = np.fft.fft(x, N)

plt.figure(figsize=(8, 2))
plt.plot(frequencies[:N//2], np.abs(x_dtft[:N//2]) / len(x))
plt.title('DTFT (Continuous Frequency Spectrum of Rectangular Function)')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

```python
X_dft = np.fft.fft(x)
frequencies_dft = np.fft.fftfreq(len(x), 1/fs)

plt.figure(figsize=(8, 2))
plt.stem(frequencies_dft[:len(x)//2], np.abs(X_dft[:len(x)//2]) / len(x), use_line_collection=True)
plt.title('DFT (Discrete Frequency Spectrum of Rectangular Function)')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

 The DTFT provides a complete, continuous representation of a signal's frequency content, while the DFT offers a discrete approximation that is limited to specific frequency samples.

Sinusoidal Signals:
A pure sine wave has only one frequency. In the frequency domain (using DTFT or DFT), it appears as a sharp spike at that specific frequency, meaning all its energy is concentrated there.

Composite Signals:
When you mix multiple sine waves at different frequencies, the frequency domain shows several spikes. Each spike represents one of the sine waves, making it easy to identify the frequencies present in the signal.

Exponentially Decaying Signals:
For signals that gradually fade over time, like an exponentially decaying signal, the frequency domain shows a spread-out spectrum. If the signal fades faster, the spread is wider, meaning more frequencies are involved. This happens because rapid changes in the time domain affect a broader range of frequencies.

Rectangular Signals:
A rectangular pulse (a signal that switches on and off abruptly) produces a frequency pattern that looks like a sinc wave, with a central peak and side ripples. The sharper the on-off transitions in the time domain, the more spread out the frequencies become. Faster transitions involve more frequencies.
